```python
import pandas as pd
# labelled = pd.read_csv("AsianReligionsData/AllBooks_baseline_DTM_Labelled.csv")
unlabelled = pd.read_csv("AsianReligionsData/AllBooks_baseline_DTM_Unlabelled.csv")

unlabelled.head()
```

```python
# Czestotliwosc zamiast liczby
columns = list(unlabelled.columns)
unlabelled['sum'] = unlabelled.sum(axis=1)

for column in columns:
    unlabelled[column] = unlabelled.apply(lambda row: row[column]/row['sum'], axis=1)

unlabelled.head()
```

```python
mean = unlabelled.loc[:, unlabelled.columns != 'sum'].mean()
mean.describe()
```

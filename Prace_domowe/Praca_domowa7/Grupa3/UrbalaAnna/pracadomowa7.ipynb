```python
from sklearn.decomposition import PCA
import matplotlib.image as image
import pandas as pd
import matplotlib.pyplot as plt
```

Wczytanie obrazka:

```python
img = image.imread('mem.png')
plt.axis('off')
plt.imshow(img)
```

```python
import sys

print(img.shape)
print(sys.getsizeof(img))
```

Rozmiar macierzy _odpowiedzialnej_ za powyszy obrazek to $450*491*3=662850$ (dla pythona way ok. 2.5MB)

```python
import numpy as np

img_ = np.reshape(img, (450, 491 * 3))
img_.shape
```

```python
pca = PCA(.99).fit(img_)
img_after_pca = pca.transform(img_)
print(img_after_pca.shape)
print(sys.getsizeof(img_after_pca))
```

Dziki PCA udao nam si zredukowa macierz obrazka do rozmiaru $450*69=31050$ (ok. 120kB) czyli **o rzd wielkoci**.

```python
tmp = pca.inverse_transform(img_after_pca)
print(tmp.shape)

tmp = np.reshape(tmp, (450,491,3)) 
print(tmp.shape)
```

```python
plt.axis('off')
plt.imshow(tmp)
plt.savefig('after_pca.png')
```

Zyski:

- rozmiar macierzy mniejszy o rzd wielkoci
- rozmiar faktycznego wyjciowego obrazka mniejszy o poow
- mimo utraty jakoci mem dalej zabawny

Straty:

- rozdzielczo obrazka to $432x288$ zamiast $491x450$
- sabsza jako obrazka
- zapisany obrazek ma bia ramk, ale to pewnie wina pyplota, a nie chce mi si szuka rozwizania
